```python
data=subset
```

```python
data
```

```python
# Summing the ByteCount and PacketCount columns across the rows for internet traffic (inbound and outbound)
# This helps simplify our analysis for detecting spikes in traffic
internet_byte_columns = [col for col in data.columns if 'ByteCount' in col and 'Internet' in col]
internet_packet_columns = [col for col in data.columns if 'PacketCount' in col and 'Internet' in col]

data['TotalInternetByteCount'] = data[internet_byte_columns].sum(axis=1)
data['TotalInternetPacketCount'] = data[internet_packet_columns].sum(axis=1)

# Plotting ByteCount and PacketCount over time
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot total internet byte count
ax[0].plot(data.index, data['TotalInternetByteCount'], label='Total Byte Count', linewidth=1)
ax[0].set_title('Total Internet Byte Count Over Time')
ax[0].set_ylabel('Bytes')
ax[0].legend()

# Plot total internet packet count
ax[1].plot(data.index, data['TotalInternetPacketCount'], label='Total Packet Count', color='orange', linewidth=1)
ax[1].set_title('Total Internet Packet Count Over Time')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Packets')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate moving averages and standard deviations for byte and packet counts
# min_periods = 1 so that we don't encounter errors when we start from the first to the 29th data points.
# window_size = 50
# window_size = 40
window_size = 30
# window_size = 20
# window_size = 15
# window_size = 10 the window size is too short for the algorithm to detect outliers
data['Byte_MA'] = data['TotalInternetByteCount'].rolling(window=window_size, min_periods=1).mean()
data['Byte_STD'] = data['TotalInternetByteCount'].rolling(window=window_size, min_periods=1).std()
data['Packet_MA'] = data['TotalInternetPacketCount'].rolling(window=window_size, min_periods=1).mean()
data['Packet_STD'] = data['TotalInternetPacketCount'].rolling(window=window_size, min_periods=1).std()

# Flag anomalies where traffic exceeds 3 standard deviations from the moving average
# the ">" sign means we are focusing at DDoS attack which is classified by overwhelming traffic
deviation = 3
#deviation = 3.5
byte_anomalies = data['TotalInternetByteCount'] > (data['Byte_MA'] + deviation * data['Byte_STD'])
packet_anomalies = data['TotalInternetPacketCount'] > (data['Packet_MA'] + deviation * data['Packet_STD'])

# Combine anomalies into a single dataframe
data['Byte_Anomaly'] = byte_anomalies
data['Packet_Anomaly'] = packet_anomalies

# Extract the anomalies for visualization
anomaly_data = data[(data['Byte_Anomaly'] | data['Packet_Anomaly'])]

# Displaying the anomalies for review
anomaly_data[['TotalInternetByteCount', 'TotalInternetPacketCount', 'Byte_Anomaly', 'Packet_Anomaly']]
```

```python
# Plotting ByteCount and PacketCount over time
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot total internet byte count
ax[0].plot(anomaly_data.index, anomaly_data['TotalInternetByteCount'], label='Total Byte Count', linewidth=1)
ax[0].set_title('Anomaly Bytes over Time')
ax[0].set_ylabel('Bytes')
ax[0].legend()

# Plot total internet packet count
ax[1].plot(anomaly_data.index, anomaly_data['TotalInternetPacketCount'], label='Total Packet Count', color='orange', linewidth=1)
ax[1].set_title('Anomaly Packets over Time')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Packets')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# First class DDoS
floods_of_small_packets = anomaly_data[(anomaly_data['Byte_Anomaly'] == False) & (anomaly_data['Packet_Anomaly'] == True)]

# Plotting ByteCount and PacketCount over time
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot total internet byte count
ax[0].plot(floods_of_small_packets.index, floods_of_small_packets['TotalInternetByteCount'], label='Total Byte Count', linewidth=1)
ax[0].set_title('Anomaly Bytes over Time')
ax[0].set_ylabel('Bytes')
ax[0].legend()

# Plot total internet packet count
ax[1].plot(floods_of_small_packets.index, floods_of_small_packets['TotalInternetPacketCount'], label='Total Packet Count', color='orange', linewidth=1)
ax[1].set_title('Anomaly Packets over Time')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Packets')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Second class DDoS
large_packet_bursts = anomaly_data[(anomaly_data['Byte_Anomaly'] == True) & (anomaly_data['Packet_Anomaly'] == False)]

# Plotting ByteCount and PacketCount over time
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot total internet byte count
ax[0].plot(large_packet_bursts.index, large_packet_bursts['TotalInternetByteCount'], label='Total Byte Count', linewidth=1)
ax[0].set_title('Anomaly Bytes over Time')
ax[0].set_ylabel('Bytes')
ax[0].legend()

# Plot total internet packet count
ax[1].plot(large_packet_bursts.index, large_packet_bursts['TotalInternetPacketCount'], label='Total Packet Count', color='orange', linewidth=1)
ax[1].set_title('Anomaly Packets over Time')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Packets')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Plotting Total Internet Byte and Packet Counts with All Anomalies Highlighted
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot total internet byte count with all types of anomalies highlighted
ax[0].plot(data.index, data['TotalInternetByteCount'], label='Total Byte Count', color='orange', linewidth=1)
ax[0].scatter(data.index[data['Byte_Anomaly']], data['TotalInternetByteCount'][data['Byte_Anomaly']], 
              color='red', label='Anomaly (Byte)', marker='o', s=50)
ax[0].scatter(data.index[(data['Byte_Anomaly'] == True) & (data['Packet_Anomaly'] == False)],
              data['TotalInternetByteCount'][(data['Byte_Anomaly'] == True) & (data['Packet_Anomaly'] == False)],
              color='blue', label='Byte Anomaly Only', marker='x', s=50)
ax[0].set_title('Total Internet Byte Count with Anomalies')
ax[0].set_ylabel('Bytes')
ax[0].legend()

# Plot total internet packet count with all types of anomalies highlighted
ax[1].plot(data.index, data['TotalInternetPacketCount'], label='Total Packet Count', color='orange', linewidth=1)
ax[1].scatter(data.index[data['Packet_Anomaly']], data['TotalInternetPacketCount'][data['Packet_Anomaly']], 
              color='red', label='Anomaly (Packet)', marker='o', s=50)
ax[1].scatter(data.index[(data['Byte_Anomaly'] == False) & (data['Packet_Anomaly'] == True)],
              data['TotalInternetPacketCount'][(data['Byte_Anomaly'] == False) & (data['Packet_Anomaly'] == True)],
              color='green', label='Packet Anomaly Only', marker='x', s=50)
ax[1].set_title('Total Internet Packet Count with Anomalies')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Packets')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
# Dropping columns with over 99.9% zeros
columns_to_drop = ['FromInternetTCPPort443IPurn:ietf:params:mud:controllerByteCount',
                   'FromInternetTCPPort443IPurn:ietf:params:mud:controllerPacketCount',
                   'ToInternetTCPPort443IPurn:ietf:params:mud:controllerByteCount',
                   'ToInternetTCPPort443IPurn:ietf:params:mud:controllerPacketCount',
                   'ToLocalUDPPort5353IP224.0.0.251/32ByteCount',
                   'ToLocalUDPPort5353IP224.0.0.251/32PacketCount']

# Dropping specified columns
data_reduced = data.drop(columns=columns_to_drop)

# Selecting relevant features and standardizing them, time is index
X_reduced = data_reduced.drop(['device_mac'], axis=1, errors='ignore')
scaler = StandardScaler()
X_reduced_scaled = scaler.fit_transform(X_reduced)

# Applying isolation forest
isolation_forest_reduced = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
data_reduced['anomaly_score'] = isolation_forest_reduced.fit_predict(X_reduced_scaled)
data_reduced['is_anomaly'] = data_reduced['anomaly_score'] == -1  # Mark anomalies where score is -1

anomalies_reduced_detected = data_reduced[data_reduced['is_anomaly']]
```

```python
# Define the relevant columns for Isolation Forest
byte_columns = [col for col in data_reduced.columns if 'ByteCount' in col]
packet_columns = [col for col in data_reduced.columns if 'PacketCount' in col]

# Define a threshold to filter out low-traffic anomalies, picked from observations above
byte_threshold = 20000
packet_threshold = 50

data_reduced['total_byte_count'] = data_reduced[byte_columns].sum(axis=1)
data_reduced['total_packet_count'] = data_reduced[packet_columns].sum(axis=1)

# Apply post-filtering to keep only high-traffic anomalies
prefiltering = data_reduced[data_reduced['is_anomaly']]
postfiltering = prefiltering[(prefiltering['total_byte_count'] > byte_threshold) 
                             | (prefiltering['total_packet_count'] > packet_threshold)]

# Display the filtered anomalies
print("High-Traffic Anomalies Detected:")
print(len(postfiltering))
```
